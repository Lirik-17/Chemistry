```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Генерируем случайные данные
np.random.seed(42)
x = np.linspace(0, 10, 50)
y = np.sin(x) + np.random.normal(scale=0.1, size=len(x))  # Исходный ряд
orig = np.sin(x)

# Применяем экспоненциальное сглаживание
alpha = 0.1  # Коэффициент сглаживания
model = SimpleExpSmoothing(y).fit(smoothing_level=alpha, optimized=False)
y_smoothed = model.fittedvalues

# Визуализация
plt.scatter(x, y, label="Оригинальный ряд")
plt.plot(x, y_smoothed, 'r-', label=f"Сглаженный ряд (α={alpha})", linewidth=2)
plt.plot(x, orig, 'r-', label=f"Сглаженный ряд (α={alpha})", linewidth=2)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import Rbf

# Генерируем случайные точки
np.random.seed(2)

x = np.linspace(0, 10, 40)
y = np.sin(x) + np.random.normal(scale=0.2, size=len(x))  # Исходные точки с шумом

# Интерполяция
rbf = Rbf(x, y, function='cubic')  # Используем мультиквадратичную функцию
x_new = np.linspace(0, 10, 100)
y_new = rbf(x_new)

# Визуализация
plt.scatter(x, y, label="Исходные точки", color="red")
plt.plot(x_new, y_new, label="RBF-интерполяция", color="blue")
plt.legend()
plt.show()
```
