```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

#   
np.random.seed(42)
x = np.linspace(0, 10, 50)
y = np.sin(x) + np.random.normal(scale=0.1, size=len(x))  #  
orig = np.sin(x)

#   
alpha = 0.1  #  
model = SimpleExpSmoothing(y).fit(smoothing_level=alpha, optimized=False)
y_smoothed = model.fittedvalues

# 
plt.scatter(x, y, label=" ")
plt.plot(x, y_smoothed, 'r-', label=f"  (={alpha})", linewidth=2)
plt.plot(x, orig, 'r-', label=f"  (={alpha})", linewidth=2)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import Rbf

#   
np.random.seed(2)

x = np.linspace(0, 10, 40)
y = np.sin(x) + np.random.normal(scale=0.2, size=len(x))  #    

# 
rbf = Rbf(x, y, function='cubic')  #   
x_new = np.linspace(0, 10, 100)
y_new = rbf(x_new)

# 
plt.scatter(x, y, label=" ", color="red")
plt.plot(x_new, y_new, label="RBF-", color="blue")
plt.legend()
plt.show()
```
